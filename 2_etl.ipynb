```python
# Import libararies
import pandas as pd
import numpy as np
```

```python
# Read data
df = pd.read_csv("performance_data_pipeline.csv")
```

```python
# Display data
df
```

```python
# Check data size
df.shape
```

```python
# Check data types
df.dtypes
```

```python
# Verify data types after converting
df.info()
```

```python
# Convert datatypes from object to datetime
df["DateCreated"] = pd.to_datetime(df["DateCreated"])
```

```python
# Convert datatypes from object to datetime
df["DateClosed"] = pd.to_datetime(df["DateClosed"])
```

```python
# Verify datatypes after converting
df.dtypes
```

```python
# Check duplicates
df.duplicated().sum()

# Remove df = df.drop_duplicates()
```

```python
# Check missing values - do not need to remove due to open case ids
df.isnull().sum()
```

```python
# Fill SLA_Met column with pending
df["SLA_Met"] = df["SLA_Met"].fillna("Pending")

# Check SLA_Met column after filled with pending
df["SLA_Met"].value_counts(dropna=False)
```

```python
# Fill CSAT column with 0
df["CSAT"] = df["CSAT"].fillna(0)

# Check CSAT column after filled with 0
df[["Status", "CSAT"]].head(10)
```

```python
# Find resolution days
df["ResolutionDays"] = (
    df["ResolutionHours"] / 24
).round(2)
```

```python
# Create resolution days
df["ResolutionDays"] = (
    df["DateClosed"] - df["DateCreated"]
).dt.days


# Check columns
df[["DateCreated","DateClosed","ResolutionDays"]].head()
```

```python
# Create month
df["Month"] = df["DateCreated"].dt.month_name()

# Check column
df["Month"].head()
```

```python
# Create year
df["Year"] = df["DateCreated"].dt.year

# Check column
df["Year"].unique()
```

```python
# Create day name
df["Day"] = df["DateCreated"].dt.day_name()

# Check column
df[["DateCreated", "Day"]].head()
```

```python
# Create week
df["Week"] = df["DateCreated"].dt.isocalendar().week

# Check column
df[["DateCreated","Week"]].head()
```

```python
# Check new columns
df.head()
```

```python
# Create new columns
df["Month"] = df["DateCreated"].dt.month_name()
```

```python
# Create new columns
df["Year"] = df["DateCreated"].dt.year
```

```python
# Create case age
today = pd.Timestamp.today()

df["CaseAge"] = np.where(

    df["Status"]=="Closed",

    (df["DateClosed"]-df["DateCreated"]).dt.days,

    (today-df["DateCreated"]).dt.days

)
```

```python
# Create SLA %
df["SLA_Met_Flag"] = (
    df["SLA_Met"]
      .map({
          "Yes":1,
          "No":0
      })
)
```

```python
# Create CSAT category
df["CSAT_Category"] = pd.cut(

    df["CSAT"],

    bins=[0,4,4.5,5],

    labels=[

        "Low",

        "Medium",

        "High"

    ]

)
```

```python
# Create resolution category
df["Resolution_Category"] = pd.cut(

    df["ResolutionHours"],

    bins=[0,24,48,72,100],

    labels=[

        "<24 hrs",

        "24-48 hrs",

        "49-72 hrs",

        ">72 hrs"

    ]

)
```

```python
# Create team KPI summary
df["Resolution_Category"] = pd.cut(

    df["ResolutionHours"],

    bins=[0,24,48,72,100],

    labels=[

        "<24 hrs",

        "24-48 hrs",

        "49-72 hrs",

        ">72 hrs"

    ]

)
```

```python
# Create officer KPI summary
officer_summary = (

    df

    .groupby("OfficerName")

    .agg(

        Cases=("CaseID","count"),

        AvgResolution=("ResolutionHours","mean"),

        AvgCSAT=("CSAT","mean")

    )

)

officer_summary
```
